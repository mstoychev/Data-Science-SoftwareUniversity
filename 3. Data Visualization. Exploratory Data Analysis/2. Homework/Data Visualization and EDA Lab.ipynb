```python
def most_prominent_marital_status(dataframe):

    maritial_status = dataframe.marital_status.value_counts()
    status = maritial_status.index[0]
    num_people = maritial_status[0]
    
    return (status, num_people)


most_prominent_marital_status(income_data)
```

Let's get another view of the data. Split the ages into three:
1. Young people: $\text{age} \le 30$
2. Middle-aged people: $30 < \text{age} \le 60$
3. Old people: $60 < \text{age}$

Return the counts in the following function. Which age group has the most people? How many are there?

```python
income_data[(income_data.age > 30) & (income_data.age <= 60)]
```

```python
def get_num_people_by_age_category(dataframe):
    young, middle_aged, old = (0, 0, 0)
    
    young = dataframe[dataframe.age <= 30].age.count()
    middle_aged = dataframe[(dataframe.age > 30) & (dataframe.age <= 60)].age.count()
    old = dataframe[dataframe.age > 60].age.count()

    return young, middle_aged, old


get_num_people_by_age_category(income_data)
```

```python
young, middle_aged, old = get_num_people_by_age_category(income_data)
nose.tools.assert_greater(young, 0)
nose.tools.assert_greater(middle_aged, 0)
nose.tools.assert_greater(old, 0)
```

Now we can create a bar chart. Execute the code below to see it.

```python
plt.title("Distribution of people by age groups")
plt.bar(range(3), [young, middle_aged, old])
plt.xticks(range(3), ["Young", "Middle-aged", "Old"])
plt.ylabel("Count")
plt.show()
```

### Problem 6. Native country (2 points)
Have a look at the native country of the people. The highest number of people are, as expected, from the US. What country makes for **the second highest** number of people? How many are they? Write a function to **calculate** and return the answer given a dataframe. DO NOT hardcode the answer, e.g. `return "Germany"`.

```python
def get_second_highest_num_people(dataframe):
    native_counts = dataframe.native_country.value_counts()
    num_people, country = native_counts[1], native_counts.index[1]
    
    return num_people, country


num_people, country = get_second_highest_num_people(income_data)
print(country)
print(num_people)
```

### Problem 7. Busiest occupations (2 points)
Which people are most overworked? Group all data by occupation and calculate the mean hours per week for each group.

Write a function that **calculates and returns** all mean hours per week as a `Series`. Sort the results in descending order (most hours to fewest hours).

```python
def get_mean_working_hours_by_occupation(dataframe):
    mean_sorted = dataframe.groupby("occupation")["hours_per_week"].mean().sort_values(ascending=False)
    
    return mean_sorted
```

```python
hours = get_mean_working_hours_by_occupation(income_data)
print(hours)
nose.tools.assert_almost_equal(hours["Handlers-cleaners"], 37.95, delta = 0.01)
```

Finally, let's plot a bar chart. Check the values carefully. If your do not match, feel free to edit the chart generation code below.

```python
plt.figure(figsize = (12, 6))
plt.barh(range(len(hours)), hours)
plt.yticks(list(range(len(hours))), hours.index)
plt.xlabel("average hours per week")
plt.title("Weekly hours by occupation", size = "x-large")
plt.show()
```
